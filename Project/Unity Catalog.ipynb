```python
%sql
use datamaster.bronze;
drop table if exists customer_reviews;
create or replace table customer_reviews(
  customer_id int,
  first_name string,
  last_name string,
  review string
);

INSERT into customer_reviews values(1,'John','Smith','Great product!'),(2,'Jane','Doe','Not so great'),(3,'Bob','Johnson','Excellent!');
select * from customer_reviews;
```

```python
%sql
create function if not exists datamaster.bronze.myname(first_name string,last_name string) returns string
  return concat_ws(' ',first_name,last_name)
  ;
select *,datamaster.bronze.myname(first_name,last_name) from datamaster.bronze.customer_reviews;
```

```python
%sql
create function datamaster.bronze.reviews(review string) returns string
  return 
  case 
    when review like '%bad%' or review like '%Not good%' then 'Negative'
    when review like '%good%' or review like '%Excellent%' then 'Positive'
  end 
  ;
select *,datamaster.bronze.myname(first_name,last_name) as full_name,datamaster.bronze.reviews(review) as review from datamaster.bronze.customer_reviews;
```
