iii)create using the read file function


```python
#/Volumes/datamaster/bronze/customer_raw_file
df=spark.read.csv('/Volumes/datamaster/bronze/customer_raw_file',header=True,inferSchema=True)
df.write.mode('overwrite').saveAsTable('datamaster.bronze.customer')

```

```python
%sql
CREATE TABLE datamaster.bronze.customer_ctas
select *,current_timestamp() as ingested_time from read_files('/Volumes/datamaster/bronze/customer_raw_file', format => 'csv')
```
